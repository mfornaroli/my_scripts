```python
from matplotlib import pyplot as plt
# 
# median salary for developers
import numpy as np
plt.style.use('classic')
import pandas as pd
df = pd.read_csv(r"https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/05-Fill_Betweens/data.csv")
print(df.columns)
dev_salaries = df['All_Devs']
py_salaries = df['Python']
java_salaries = df['JavaScript']
ages = df['Age']
overall_median = 57287

plt.plot(ages, dev_salaries, color = 'k', linestyle= '--', marker='o',label='All devs')
plt.plot(ages, py_salaries, color = 'b', label="Python", linewidth=3)

plt.title('median salary (USD) by age')
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")
# new function useful for filling up areas in graphs
plt.fill_between(ages, 
                 py_salaries, 
                 dev_salaries, 
                 where=(py_salaries>=dev_salaries),
                 alpha=0.3,
                 color='yellow',    
                 label='Above avg'
                )
plt.fill_between(ages, 
                 py_salaries, 
                 dev_salaries, 
                 where=(py_salaries<dev_salaries),
                 alpha=0.5,
                 color='red',
                 label='Below avg'
                )

plt.legend(loc='upper left', fontsize='small')
plt.grid(True)
```

```python
import matplotlib.pyplot as plt

# Start with one
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([1,2,3])

# Now later you get a new subplot; change the geometry of the existing
n = len(fig.axes)
for i in range(n):
    fig.axes[i].change_geometry(n+1, 1, i+1)

# Add the new
ax = fig.add_subplot(n+1, 1, n+1)
ax.plot([4,5,6])

plt.show() 
```
